```python
import sympy as sp
import numpy as np
from IPython.display import display, Math
import warnings
warnings.filterwarnings("ignore")

sp.init_printing(use_latex='mathjax')
t, f = sp.symbols('t f', real=True)
pi = sp.pi
j_sym = sp.Symbol('j') # Símbolo seguro para la j

def traductor_dios(tramos):
    print("==================================================")
    print(" 🤖 TRADUCTOR DIOS: AUTO-SIMETRÍA ACTIVADA")
    print("==================================================\n")
    
    # 1. Extraer propiedades físicas de todas las piezas
    piezas = []
    for t in tramos:
        tipo = t[0]
        if tipo == 'rect':
            tau = t[3] - t[2]
            centro = (t[3] + t[2]) / 2
            area = t[1] * tau
            piezas.append({'tipo': tipo, 'area': area, 'tau': tau, 'centro': centro, 'usado': False})
        elif tipo == 'tri':
            base = t[3] - t[2]
            centro = (t[3] + t[2]) / 2
            area = (t[1] * base) / 2
            piezas.append({'tipo': tipo, 'area': area, 'tau': base, 'centro': centro, 'usado': False})
    
    respuestas = []
    
    # 2. El Toque Humano: Buscar parejas simétricas para aplicar Euler
    for i, p1 in enumerate(piezas):
        if p1['usado']: continue
        
        pareja_encontrada = False
        if p1['centro'] != 0:
            for j, p2 in enumerate(piezas):
                # Si hay otra pieza igualita, en el lado opuesto del plano
                if i != j and not p2['usado'] and p1['tipo'] == p2['tipo'] and p1['area'] == p2['area'] and p1['tau'] == p2['tau']:
                    if abs(p1['centro'] + p2['centro']) < 0.0001: 
                        pareja_encontrada = True
                        p1['usado'] = True
                        p2['usado'] = True
                        
                        c_val = abs(p1['centro'])
                        # Fusionar en un Coseno! (Área x 2)
                        if p1['tipo'] == 'rect':
                            respuestas.append(f"[{2 * p1['area']} * sinc({p1['tau']}πf) * cos({2 * c_val}πf)]")
                        else: # tri
                            respuestas.append(f"[{2 * p1['area']} * sinc²({p1['tau']/2}πf) * cos({2 * c_val}πf)]")
                        break
        
        # 3. Si no tiene pareja (o está justo en el centro), se queda normal
        if not pareja_encontrada:
            p1['usado'] = True
            c_val = p1['centro']
            exponencial = "" if c_val == 0 else f" * e^(-j{2*c_val}πf)"
            
            if p1['tipo'] == 'rect':
                respuestas.append(f"[{p1['area']} * sinc({p1['tau']}πf){exponencial}]")
            else: # tri
                respuestas.append(f"[{p1['area']} * sinc²({p1['tau']/2}πf){exponencial}]")

    print("Respuesta Final de 100 Puntos: X(f) = \n" + "  +  \n".join(respuestas))
    print("\n==================================================")



# ==========================================
# 📝 ZONA DE CONFIGURACIÓN VISUAL
# ==========================================
# Ejemplos de cómo traducir el dibujo del examen:

# Si viene un pulso rectangular de amplitud 3, de 0 a 4:
# mis_tramos = [('rect', 3, 0, 4)]

# Si viene un triángulo de amplitud 5, centrado de -2 a 2:
# mis_tramos = [('tri', 5, -2, 2)]

# Si viene una flecha (impulso) de amplitud 10, ubicada en t = 3:
# mis_tramos = [('dirac', 10, 3)]

# ==========================================
# PRUEBA DEL EXAMEN (El Podio Completo)
# ==========================================
mis_tramos = [('rect', 2, -2, -1), ('rect', 1, -1, 1), ('rect', 2, 1, 2)]
traductor_dios(mis_tramos)
```
