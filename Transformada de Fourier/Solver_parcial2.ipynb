```python
import sympy as sp
import numpy as np
from IPython.display import display, Math
import warnings
warnings.filterwarnings("ignore")

sp.init_printing(use_latex='mathjax')
t, f = sp.symbols('t f', real=True)
pi = sp.pi
j_sym = sp.Symbol('j') # Símbolo seguro para la j

def traductor_maestro(tramos):
    print("==================================================")
    print(" 👑 TRADUCTOR MAESTRO (RECTÁNGULOS, TRIÁNGULOS Y DIRACS)")
    print("==================================================\n")
    
    respuestas = []
    
    for tramo in tramos:
        tipo = tramo[0]
        
        # 1. BLOQUE RECTANGULAR -> Genera Sinc
        if tipo == 'rect':
            _, A, t_ini, t_fin = tramo
            tau = t_fin - t_ini
            t_centro = (t_fin + t_ini) / 2
            area = A * tau
            
            if t_centro == 0:
                respuestas.append(f"[{area} * sinc({tau}πf)]")
            else:
                respuestas.append(f"[{area} * sinc({tau}πf) * e^(-j{2*t_centro}πf)]")
                
        # 2. BLOQUE TRIANGULAR -> Genera Sinc Cuadrado
        elif tipo == 'tri':
            _, A, t_ini, t_fin = tramo
            base = t_fin - t_ini
            t_centro = (t_fin + t_ini) / 2
            area = (A * base) / 2
            
            if t_centro == 0:
                respuestas.append(f"[{area} * sinc²({base/2}πf)]")
            else:
                respuestas.append(f"[{area} * sinc²({base/2}πf) * e^(-j{2*t_centro}πf)]")
                
        # 3. IMPULSO (FLECHA) -> Genera Constante o Exponencial Compleja
        elif tipo == 'dirac':
            # El formato es: ('dirac', Area_de_flecha, posicion_en_el_tiempo)
            _, A, t_pos = tramo
            
            if t_pos == 0:
                respuestas.append(f"[{A}]")
            else:
                respuestas.append(f"[{A} * e^(-j{2*t_pos}πf)]")

    print("Respuesta Final: X(f) = \n" + "  +  \n".join(respuestas))
    print("\n==================================================")

# ==========================================
# 📝 ZONA DE CONFIGURACIÓN VISUAL
# ==========================================
# Ejemplos de cómo traducir el dibujo del examen:

# Si viene un pulso rectangular de amplitud 3, de 0 a 4:
# mis_tramos = [('rect', 3, 0, 4)]

# Si viene un triángulo de amplitud 5, centrado de -2 a 2:
# mis_tramos = [('tri', 5, -2, 2)]

# Si viene una flecha (impulso) de amplitud 10, ubicada en t = 3:
# mis_tramos = [('dirac', 10, 3)]

# PRUEBA ACTIVA:
mis_tramos = [('tri', 5, -2, 2)]
traductor_maestro(mis_tramos)
```
