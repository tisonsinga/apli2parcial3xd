```python
import sympy as sp
from IPython.display import display, Math

sp.init_printing(use_latex='mathjax')
t, w = sp.symbols('t w', real=True)
j = sp.I

def deconstructor_procedimiento_universal(F_w_input, f_t_input):
    print("==================================================")
    print(" 🛠️ DECONSTRUCTOR DE PROCEDIMIENTO (UNIVERSAL)")
    print("==================================================\n")

    # 1. Detectar w0 desde f(t) (buscando e^(j*w0*t))
    w0 = 0
    f_base_t = f_t_input
    
    # Buscamos términos exponenciales complejos
    for factor in sp.preorder_traversal(f_t_input):
        if factor.is_Pow and factor.base == sp.E and factor.exp.has(j):
            # Extraemos lo que multiplica a j*t
            w0 = sp.simplify(factor.exp / (j * t))
            f_base_t = sp.simplify(f_t_input / factor)
            break

    print("PASO 1: Identificación de la Señal Base")
    print("Al analizar la respuesta, detectamos la función base sin el corrimiento:")
    display(Math(rf"f_{{base}}(t) = {sp.latex(f_base_t)}"))
    
    # 2. Reconstruir la F_base(w)
    print("\nPASO 2: Transformada Base de Tabla")
    F_base_w = sp.simplify(F_w_input.subs(w, w + w0))
    print("Por inspección en tablas, esta f_base(t) corresponde a:")
    display(Math(rf"F_{{base}}(\omega) = {sp.latex(F_base_w)}"))

    # 3. Justificación de Álgebra (Fracciones Parciales)
    if sp.denom(F_base_w) != 1:
        print("\nPASO 3: Desarrollo Algebraico")
        print("Se aplica expansión en fracciones parciales sobre la base para facilitar la inversión:")
        display(Math(rf"F_{{base}}(\omega) = {sp.latex(sp.apart(F_base_w, w))}"))

    # 4. Aplicar Teorema
    print("\nPASO 4: Aplicación del Teorema de Corrimiento")
    if w0 != 0:
        print(f"Dado que la función original F(w) presenta un desplazamiento de w0 = {w0},")
        print(f"aplicamos el Teorema de Corrimiento en Frecuencia: f(t) -> F(w - w0) implica multiplicar por e^(j*w0*t)")
        display(Math(rf"\mathfrak{{F}}^{{-1}}\{{ F(\omega - {sp.latex(w0)}) \}} = f_{{base}}(t) \cdot e^{{j ({sp.latex(w0)}) t}}"))
    else:
        print("No se detectaron corrimientos, la inversión es directa por tabla.")

    print("\n==================================================")
    print(" RESULTADO FINAL PARA EL EXAMEN:")
    display(Math(rf"f(t) = {sp.latex(f_t_input)}"))
    print("==================================================")

# --- EJEMPLO DE USO (Problema del Rompecabezas) ---
F_examen = 4/((1+j*w)*(3+j*w))
# Aquí pegas la respuesta que te dio la Caja Negra (ya simplificada o con sinh)
f_caja_negra = (2-2*sp.exp(-2*t))*sp.exp(-t)

deconstructor_procedimiento_universal(F_examen, f_caja_negra)
```
