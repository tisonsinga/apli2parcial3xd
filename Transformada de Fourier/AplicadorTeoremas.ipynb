## NOTA TEÓRICA: El Signo del Corrimiento ($w_0$)

La trampa más común al usar el Teorema de Corrimiento en Frecuencia es equivocarse en el signo de la variable `w0` al meterla al script. Todo se reduce a **comparar patrones**.

### 1. La Fórmula "Plantilla" (El Estándar)
El teorema dicta que si multiplicas tu función por una exponencial imaginaria **positiva**, la frecuencia se resta:
$$f(t) e^{i \omega_0 t} \xrightarrow{\mathfrak{F}} F(\omega - \omega_0)$$

> *Nota: El exponente de la plantilla siempre es $+i \omega_0 t$*

### 2. El Problema del Examen (El Caso Real)
Supongamos que el ingeniero te pone la siguiente función:
$$g(t) = e^{-i \alpha t} (1 - t^2)$$

### 3. El Método de Igualación
Para saber exactamente qué valor darle a la variable `w0` en tu código, debes igualar el exponente de la plantilla con el exponente de tu problema:

$$i \omega_0 t = -i \alpha t$$

Si cancelamos la unidad imaginaria ($i$) y el tiempo ($t$) de ambos lados, descubrimos la verdad:
$$\omega_0 = -\alpha$$

### ⚠️ ¿Por qué importa esto para el Script?
El código está programado para sustituir estrictamente la fórmula teórica: $F(\omega - \omega_0)$.
* **Si lo haces mal (Pones $w_0 = \alpha$):** El script calculará $F(\omega - \alpha)$. ¡Respuesta incorrecta!
* **Si lo haces bien (Pones $w_0 = -\alpha$):** El script aplicará ley de signos: $F(\omega - (-\alpha))$, y te arrojará $F(\omega + \alpha)$. ¡Respuesta perfecta!

**Regla de Supervivencia:** Compara el exponente de tu examen con un exponente positivo imaginario. Lo que sobre, **con todo y su signo**, es el valor que debes poner en `mi_w0` en tu script.
